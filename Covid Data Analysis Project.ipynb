```python
import pandas as pd
data = {
    "state/UnionTerritory": ["Maharashtra", "Kerla", "Madhya Pradesh"],
    "Confirmed": [1000, 500, 800],
    "Deaths": [50, 20, 30],
    "Cured": [800, 400, 600]
}

statewise = pd.DataFrame(data)


statewise.set_index("state/UnionTerritory")
```

#Top 10 active cases in satetwise


---



```python
import pandas as pd
import matplotlib.pyplot as plt

statewise['Active Cases'] = statewise['Confirmed'] - statewise['Deaths'] - statewise['Cured']

# Select Top 10 States with Highest Active Cases
top_10_states = statewise.nlargest(10, 'Active Cases')

data = top_10_states[['Confirmed', 'Deaths', 'Cured', 'Active Cases']]

colors = plt.cm.BuPu(data / data.max())

# Create Matplotlib Figure and Axes
fig, ax = plt.subplots(figsize=(10, 6))
ax.axis('off')  # Hide axis

table = ax.table(cellText=data.values,colLabels=data.columns,loc='center', cellColours=colors)

table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.2, 1.2)

ax.set_title('Top 10 States with Highest Active Cases', fontsize=16, pad=20)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "State/unionTerritory": ["Maharashtra","assam", "Bihar", "Delhi",  "Karnataka",
                             "Gujarat", "Kerala", "madhyaPradesh", "Punjab", "UttarPradesh"],
    "Active_cases": [8177105,746234,855654,2042912,4096573,4096573, 3611688, 1292436,691767,1056744]
}

top_10_active_cases = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
ax = sns.barplot(data=top_10_active_cases, x="State/unionTerritory", y="Active_cases",
                 linewidth=2, edgecolor='red')

# Customize Plot
ax.set_title('Top 10 States/Union Territories with Highest Active Cases', fontsize=16)
ax.set_xlabel('State/Union Territory', fontsize=14)
ax.set_ylabel('Active Cases', fontsize=14)
ax.tick_params(axis='x', labelrotation=45)


plt.show()
```

Top states with highest deaths


```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'State/UnionTerritory': ['Maharashtra', 'Karnataka', 'Kerala','UttarPradesh', 'Gujarat',
                              'MadhyaPradesh', 'Delhi', 'Assam', 'Bihar', 'Punjab'],
    'Total Deaths': [148598, 40404, 72105, 23724, 11099, 10789, 26691, 8037, 12315, 20587]
}

top_10_states_death = pd.DataFrame(data)

top_10_states_death = top_10_states_death.sort_values(by='Total Deaths', ascending=False)

# Plotting the top 10 states with the highest death cases using Matplotlib
plt.figure(figsize=(12, 8))
plt.bar(top_10_states_death['State/UnionTerritory'], top_10_states_death['Total Deaths'],
        color='red', edgecolor='black', linewidth=1)

plt.xlabel('State/Union Territory', fontsize=12)
plt.ylabel('Total Death Cases', fontsize=12)
plt.title('Top 10 States/Union Territories with Highest Death Cases', fontsize=14)

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=10)

plt.tight_layout()
plt.show()

```

Male vs Female Vaccination


```python
import pandas as pd
import plotly.express as px
data = {
    'State/UnionTerritory': ['State1', 'State2', 'State3', 'State4', 'State5'],
    'Male(Individuals Vaccinated)': [10000, 8000, 6000, 4000, 2000],
    'Female(Individuals Vaccinated)': [9000, 7000, 5000, 3000, 1000]
}

vaccination = pd.DataFrame(data)

# Calculate the sum of vaccinated individuals for males and females
male_vaccinated = vaccination["Male(Individuals Vaccinated)"].sum()
female_vaccinated = vaccination["Female(Individuals Vaccinated)"].sum()

# Create a pie chart using Plotly
fig = px.pie(names=['Male', 'Female'], values=[male_vaccinated, female_vaccinated],
             title='Distribution of Male and Female COVID-19 Vaccination')
fig.show()
```
